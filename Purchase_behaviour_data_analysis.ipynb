```python
import pandas as pd
```

```python
df=pd.read_csv("customer_shopping_behavior.csv")
```

```python
df.head()
```

```python
df.describe(include="all")
```

```python
df.isnull().sum()
```

```python
df['Review Rating']=df.groupby('Category')['Review Rating'].transform(lambda x:x.fillna(x.median()))
```

```python
df.isnull().sum()
```

```python
df.columns=df.columns.str.lower()
df.columns=df.columns.str.replace(" ","_")
```

```python
df.columns
```

```python
df=df.rename(columns={'purchase_amount_(usd)':'purchase_amount'})
```

```python
labels=['Young Adult','Adult',"Middle-Aged","Senoir"]
df["age_group"]=pd.qcut(df['age'],q=4,labels=labels)
```

```python
df[['age','age_group']].head(20)
```

```python
df.columns
```

```python
df['frequency_of_purchases'].unique()
```

```python
freq_map={'Fortnightly':14,
          'Weekly':7,
          'Annually':365, 
          'Quarterly':90, 
          'Bi-Weekly':14,
         'Monthly':30,
           'Every 3 Monthly':90}
          
```

```python
# freq of purchases to numericals to be handle easly in analysis
df['purchase_frequency_days']=df['frequency_of_purchases'].map(freq_map)
```

```python
df[["frequency_of_purchases","purchase_frequency_days"]].tail(20)
```

```python
df[['discount_applied', 'promo_code_used']]

```

```python
df=df.drop("promo_code_used", axis=1)
```
